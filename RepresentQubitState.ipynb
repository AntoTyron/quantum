```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from math import sqrt, pi
```

```python
qc = QuantumCircuit(1)
initial_state = [0,1]   # equivalent to 1
qc.initialize(initial_state,0)   # set q0 to |1>
qc.draw('text')
```

```python
backend = Aer.get_backend('statevector_simulator')
qc = QuantumCircuit(1)
initial_state = [0,1]   # equivalent to 1
qc.initialize(initial_state,0)   # set q0 to |1>
result = execute(qc, backend).result()
out_statevector = result.get_statevector()
print(out_statevector)   # 0.+0.j=0 1.+0.j=1
count = result.get_counts()
plot_histogram(count)
```

```python
backend = Aer.get_backend('statevector_simulator')
qc = QuantumCircuit(1)
initial_state=[sqrt(2)/sqrt(6),1j*sqrt(4)/sqrt(6)]
qc.initialize(initial_state,0)
result = execute(qc,backend).result()
qc1 = QuantumCircuit(1)
initial_state1=[1/sqrt(3),1j*sqrt(2)/sqrt(3)]
qc1.initialize(initial_state1,0)
result1 = execute(qc1,backend).result()
print(result.get_counts() == result1.get_counts())
out_statevector = result.get_statevector()
print(out_statevector)
counts = result.get_counts()
qc.draw()
plot_histogram(counts)
```
